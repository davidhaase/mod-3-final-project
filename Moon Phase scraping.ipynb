```python
response = requests.get('https://api.usno.navy.mil/moon/phase?year=2018')
data = response.json()
```

```python
moonphase = data['phasedata']
```

```python
for i in range(2012,2020):
    response = requests.get('https://api.usno.navy.mil/moon/phase?year={}'.format(i))
    data = response.json()
    moonphase.append(data['phasedata'])
```

```python
moonphase[0]
```

```python
df = pd.DataFrame()
```

```python
for i in range(len(moonphase)):
    df = df.append(pd.DataFrame(moonphase[i]))
    
```
